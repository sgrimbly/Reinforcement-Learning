```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
train_score = [20.65,19,19.2,20.75,20.35,19.85,35.25,81.25,288.4]
eval_score = [9.38,9.46,9.38,15.28,9.34,116.98,65.8,161.98,200]
```
